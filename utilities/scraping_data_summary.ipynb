```python
def collect_master_data():
    try:
        activity_df = pd.read_csv(r'master_data_scraping_Result/master_activity.csv')
        villa_df = pd.read_csv(r'master_data_scraping_Result/master_villa.csv')
        restaurant_df = pd.read_csv(r'master_data_scraping_Result/master_restaurant.csv')
        hotel_df = pd.read_csv(r'master_data_scraping_Result/master_hotel.csv')
    except:
        activity_df = pd.DataFrame()
        villa_df = pd.DataFrame()
        restaurant_df = pd.DataFrame()
        hotel_df = pd.DataFrame()
    return activity_df, villa_df, restaurant_df, hotel_df
```

```python
def transform_file_to_dataframe(filename):
    df = pd.read_csv(filename)
    return df

def append_new_data_to_master_file(filename, df):
    activity_df, villa_df, restaurant_df, hotel_df = collect_master_data()
    if 'activity' in filename:
        new_df = pd.concat([activity_df,df])
        new_df = data_preprocessing(new_df)
        # new_df = pd.concat([new_df[col].sort_values().reset_index(drop=True) for col in df], axis=1, ignore_index=True)
        new_df.to_csv(r'master_data_scraping_Result/master_activity.csv', index=False)
    elif 'restaurant' in filename:
        new_df = pd.concat([restaurant_df,df])
        new_df = new_df.drop_duplicates(keep="last")
        new_df = data_preprocessing(new_df)
        # new_df = pd.concat([new_df[col].sort_values().reset_index(drop=True) for col in df], axis=1, ignore_index=True)
        new_df.to_csv(r'master_data_scraping_Result/master_restaurant.csv', index=False)
    elif 'hotel' in filename:
        new_df = pd.concat([hotel_df,df])
        new_df = new_df.drop_duplicates(keep="last")
        new_df = data_preprocessing(new_df)
        # new_df = pd.concat([new_df[col].sort_values().reset_index(drop=True) for col in df], axis=1, ignore_index=True)
        new_df.to_csv(r'master_data_scraping_Result/master_hotel.csv', index=False)
    elif 'villa' in filename:
        new_df = pd.concat([villa_df,df])
        new_df = new_df.drop_duplicates(keep="last")
        new_df = data_preprocessing(new_df)
        # new_df = pd.concat([new_df[col].sort_values().reset_index(drop=True) for col in df], axis=1, ignore_index=True)
        new_df.to_csv(r'master_data_scraping_Result/master_villa.csv', index=False)
    else:
        return pd.DataFrame()
```

```python
def data_preprocessing(df):
    # df = df.dropna(subset='title')
    # print(df['title'].head(5))
    # df['title'].str.replace(r'[^\w\s]+', '', regex=True)
    df = df.drop_duplicates(keep="last")
    return df
```

```python
# cell untuk save data 
activity_df.to_csv(r'master_data_scraping_Result/master_activity.csv', index=False)
villa_df.to_csv(r'master_data_scraping_Result/master_villa.csv', index=False)
restaurant_df.to_csv(r'master_data_scraping_Result/master_restaurant.csv', index=False)
hotel_df.to_csv(r'master_data_scraping_Result/master_hotel.csv', index=False)
```

```python
parent_dir = 'data_scraping_result'
for dir_ in os.listdir(parent_dir):
    filename = ""
    if 'csv' in dir_:
        filename = f"{parent_dir}/{dir_}"
        temp_df = transform_file_to_dataframe(filename)
        append_new_data_to_master_file(filename, temp_df)
        
    else: 
        print(os.listdir(f"{parent_dir}/{dir_}"))
        for child_dir in os.listdir(f"{parent_dir}/{dir_}"):
            # if 'csv' in child_dir:
            filename = f"{parent_dir}/{dir_}/{child_dir}"
            temp_df = transform_file_to_dataframe(filename)
            if len(temp_df.columns) <=3:
                continue
            print(filename, len(temp_df.columns))
            append_new_data_to_master_file(filename, temp_df)
            # else:
            #     continue
    # print(filename)   
```

```python
def collect_info_from_text(filename):
    with open(filename, 'r+', encoding="utf-8") as f:
        return f.read().splitlines()
```

```python
parent_dir = "history_result"
temp_data = []
for dir_ in os.listdir(parent_dir):
    filename = ""
    if 'txt' in dir_:
        filename = f"{parent_dir}/{dir_}"
        temp_data = temp_data + collect_info_from_text(filename)
    else:
        if "json" in dir_:
            continue
        for child_dir in os.listdir(f"{parent_dir}/{dir_}"):
            filename = f"{parent_dir}/{dir_}/{child_dir}"
            temp_data = temp_data + collect_info_from_text(filename)

with open('master_data_scraping_result/master_images.txt','w',encoding="utf-8") as f:
    temp_data = list(dict.fromkeys(temp_data))
    f.writelines(s + '\n' for s in temp_data)
```

```python
pd.options.display.max_rows = None
def get_most_regency_result(df):
    df['regency'] = df['regency'].str.strip(' ')
    new_df = df.groupby(['regency'])['regency'].count().sort_values(ascending=False)
    print(new_df.head(5))

def get_total_data(df):
    print(f"total data: {len(df)}")
    column_names = df.columns
    print(f"with total column is: {len(column_names)}")
    most_column_nan = {"title": "", "total": 0}
    for name in column_names:
        count_nan = df[name].isna().sum()
        if count_nan > most_column_nan['total']:
            most_column_nan['title'] = name
            most_column_nan['total'] = count_nan
    print(f"most column with null value is: {most_column_nan['title']} || with value: {most_column_nan['total']}")
        
def get_total_data_with_image(df,images_name_list):
    print(f"we got data with images: {len(df[df['title'].isin(images_name_list)])}")
    
```

```python
def analysis_dataframe():
    parent_dir = 'master_data_scraping_result'
    image_list_filename = 'master_data_scraping_result/master_images.txt'
    master_image_file = []
    with open(image_list_filename, 'r', encoding="utf-8") as f:
        master_image_file = f.read().splitlines()
    for dir_ in os.listdir(parent_dir):
        if 'csv' in dir_:
            print(f"analysis for {dir_}")
            df = pd.read_csv(f"{parent_dir}/{dir_}")
            get_total_data(df)
            get_total_data_with_image(df,master_image_file)
            get_most_regency_result(df)
            print("--"*30,'\n\n')
        elif 'txt' in dir_:
            pass
        

analysis_dataframe()
```
